```python
import pandas as pd
import numpy as np
df = pd.read_csv('/home/seyma/Downloads/gunluk-arac-saym.csv', sep = ';')
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df_clean = df.copy()
```

```python
df_clean.rename(columns={'Tarih': 'tarih', 'Sensor Adi': 'sensor_adi','X Koordinati': 'x_koordinati', 'Y Koordinati': 'y_koordinati' , 'Arac Sayisi' : 'arac_sayisi' }, inplace=True)
```

```python
df_clean.info()
```

```python
df_clean.drop('Unnamed: 5', axis=1, inplace=True)
```

```python
df_clean.info()
```

```python
df_clean = df_clean[df_clean.tarih.notnull()]
df_clean = df_clean[df_clean.sensor_adi.notnull()]
df_clean = df_clean[df_clean.x_koordinati.notnull()]
df_clean = df_clean[df_clean.y_koordinati.notnull()]
df_clean = df_clean[df_clean.arac_sayisi.notnull()]
df_clean.info()
```

```python
df_clean['tarih']= pd.to_datetime(df_clean['tarih'])
#df_clean.x_koordinati = df_clean.x_koordinati.astype(float)
#df_clean.y_koordinati = df_clean.y_koordinati.astype(float)
df_clean.sensor_adi = df_clean.sensor_adi.astype(str)
df_clean.arac_sayisi = df_clean.arac_sayisi.astype(int)
df_clean.info()
```

```python
df_clean.tail()
```

```python
df_clean['x_koordinati'] = df_clean['x_koordinati'].str.replace(',','.')
df_clean.head()
```

```python
df_clean = df_clean.drop(labels=1, axis=0)
```

```python
df_clean.head(20)
```

```python
df_clean = df_clean.reset_index()
```

```python
df_clean.info()
```

```python
df_clean.head(20)
```

```python
df_wrong = pd.Series(df_clean.x_koordinati.map(lambda calc: calc.count('.')) > 1)
#df_clean.x_koordinati.count('.')
df_wrong
```

```python
df_wrong_list = (df_wrong[df_wrong].index)
df_wrong_list
```

```python
df_clean.info()
```

```python
df_clean = df_clean.drop(df_clean.index[df_wrong_list])
```

```python
df_clean.x_koordinati = df_clean.x_koordinati.astype(float)
df_clean.head(20)
```

```python
df_clean['y_koordinati'] = df_clean['y_koordinati'].str.replace(',','.')
df_wrong_y = pd.Series(df_clean.y_koordinati.map(lambda calc: calc.count('.')) > 1)
df_wrong_list_y = (df_wrong_y[df_wrong_y].index)
df_clean = df_clean.drop(df_clean.index[df_wrong_list_y])
df_clean.y_koordinati = df_clean.y_koordinati.astype(float)
df_clean.head(20)
```

```python
#df_clean.columns = df_clean.columns.droplevel()
df_clean = df_clean.drop(labels=["index", "level_0"], axis=1)
```

```python
df_clean.head()
```

```python
df_clean.to_csv('cleaned.csv', index = False)
```
